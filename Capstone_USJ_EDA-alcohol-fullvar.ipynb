```python
# Weighted correlation function
def weighted_corr(x, y, w):
    x_mean = np.average(x, weights=w)
    y_mean = np.average(y, weights=w)
    cov_xy = np.average((x - x_mean) * (y - y_mean), weights=w)
    var_x = np.average((x - x_mean)**2, weights=w)
    var_y = np.average((y - y_mean)**2, weights=w)
    return cov_xy / np.sqrt(var_x * var_y)

corr = weighted_corr(temp['Age at First Drink'], temp['TotalDrinking'], temp['Weight'])
print(f"Weighted correlation between Age at First Drink and Total Drinks (past year): {corr:.3f}")
```

In this model, higher numbers indicate lower religious importance.
Thus, the variable LowReligiousImportance captures people with less religious commitment.

The results below incorporate this relationship properly.

1. Main Findings
Intercept (7.87, p = 0.001)

This is the expected TotalDrinking score for the reference religion group, among individuals with high religious importance (because LowReligiousImportance = 0).

Effect of Religion Group (among highly religious individuals)

Catholics (+7.14, p = 0.030)
Highly religious Catholics drink significantly more (+7.14 units) than the highly religious reference group.

Other religions (+18.15, p = 0.005)
Highly religious members of “Other” religions drink much more (+18.15 units) than the highly religious reference group.

Protestants (+0.40, p = 0.895)
No real difference from the reference group.

Effect of Religious Importance (4.01, p = 0.010)

Because higher scores = lower religious importance, this coefficient means:

People who consider religion less important drink 4 units more than those who consider religion important.

This main effect is statistically significant.

2. Interaction Effects (How religion modifies the importance effect)
Catholic × Low Religious Importance (−3.87, p = 0.043)

This means:

Catholics’ increase in drinking due to low religious importance is 3.87 units smaller than the general increase observed in other groups.

Interpretation:
Among Catholics, the relationship between low religious importance and higher drinking is weaker than expected.

Other × Low Religious Importance (−6.22, p = 0.071)

For those in “Other” religions:

The effect of low religious importance is about 6.22 units smaller than the general trend.

```python
# Weighted correlation function
def weighted_corr(x, y, w):
    """Compute weighted Pearson correlation."""
    x_mean = np.average(x, weights=w)
    y_mean = np.average(y, weights=w)
    cov_xy = np.average((x - x_mean) * (y - y_mean), weights=w)
    var_x = np.average((x - x_mean) ** 2, weights=w)
    var_y = np.average((y - y_mean) ** 2, weights=w)
    return cov_xy / np.sqrt(var_x * var_y)

# Drop missing values for all three columns
valid = df1.dropna(subset=['TotalDrinking', 'DisciplinaryIndex', 'Weight'])

# Compute weighted correlation
r_weighted = weighted_corr(valid['TotalDrinking'], valid['DisciplinaryIndex'], valid['Weight'])

print(f"\nWeighted correlation (Total Drinks vs DisciplinaryIndex): {r_weighted:.3f}")
```

```python
from sklearn.preprocessing import StandardScaler

substance_vars = [
    "EverTried_Cigarette",
    "EverSmoked_Regularly",
    "Age_StartedSmoking",       # reverse-code
    "Days_SmokedPast30",
    "CigsPerDay_Past30",
    "DrinkWithoutAdults",
    "Age_FirstDrinkNoAdults",   # reverse-code
    "DrinkFrequency_Past12mo",
    "DrinksPerOccasion_Past12mo",
    
]

# Standardize all variables first
scaler = StandardScaler()
df1[[v + "_Z" for v in substance_vars]] = scaler.fit_transform(df1[substance_vars])

# Reverse-code the age variables
df1["Age_StartedSmoking_rev_Z"] = -df1["Age_StartedSmoking_Z"]
df1["Age_FirstDrinkNoAdults_rev_Z"] = -df1["Age_FirstDrinkNoAdults_Z"]

# Build list of variables to include in final index
substance_vars_final = [
    "EverTried_Cigarette_Z",
    "EverSmoked_Regularly_Z",
    "Age_StartedSmoking_rev_Z",
    "Days_SmokedPast30_Z",
    "CigsPerDay_Past30_Z",
    "DrinkWithoutAdults_Z",
    "Age_FirstDrinkNoAdults_rev_Z",
    "DrinkFrequency_Past12mo_Z",
    "DrinksPerOccasion_Past12mo_Z",
    
]

# Create final composite index
df1["SubstanceUse_Index"] = df1[substance_vars_final].mean(axis=1)
```
